### Question 2.
- Vanishing gradient 문제는 긴 시퀀스를 모델링할 때, 비교적 멀리 떨어진 timestep에서의 gradient가 충분히 전달되지 못해 goldfish memory와 같은 현상을 유발했습니다. Vanilla RNN에서의 이러한 문제를 해결하기 위해 제시된 RNN의 변형 구조 두 가지를 예시로 들어 주세요.

### Answer 2.
정답:

LSTM

LSTM은 hidden state 외에 cell state가 있고 Forget, Input, Output gate를 가집니다. 그 gate들을 이용해 정보를 잊을지 유지할지를 선택하여 Vanishing gradient 문제를 해결합니다. 


GRU

LSTM이 더 간단해진 모델입니다. LSTM과 다르게 Reset gate, Update gate를 사용합니다. 또한 cell state, hidden state가 합쳐진 형태의 모델입니다.
