#### merge 실습을 위한 데이터 쪼개기


```python
df_x = df[['quality', 'fixed acidity', 'citric acid']].head(30)
df_y = df[['quality', 'volatile acidity', 'chlorides', 'alcohol']].head(30)
df_y = df_y[df_y.quality >= 7]
```

#### df_x와 df_y를 inner join시켜라 (quality를 기준으로)

```python
df_INNER_JOIN = pd.merge(df_x, df_y, left_on='quality', right_on='quality', how='inner')
df_INNER_JOIN
```

#### df_x에서 df_y를 left join시켜라 (quality를 기준으로)

```python
df_LEFT_JOIN = pd.merge(df_x, df_y, left_on='quality', right_on='quality', how='left')
df_LEFT_JOIN
```

1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라.
(계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
(전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

```python
dv = pd.read_csv('/content/gdrive/MyDrive/와이빅타/교육세션/submit/ybigta-edu/0720/deliver.csv.csv의 사본') #wine.csv도 '내드라이브'에 사본 저장 후, 경로 수정
dv.head()
```

```python
def func(month):
    if month in ["03","04","05"]:
        return '봄'
    elif month in ["06","07","08"]:
        return '여름'
    elif month in ["09","10","11"]:
        return '가을'
    elif month in ["12","01","02"]:
        return '겨을'
    else:
        return '몰라'
```

```python
dv['월']= dv['날짜'].apply(lambda x: x[5:7])
dv['계절']=dv['월'].apply(func)
dv.head(30)
```

```python
dv['미세먼지(㎍/㎥)'].isnull().sum()
```

```python
dv['미세먼지(㎍/㎥)'].fillna(dv['미세먼지(㎍/㎥)'].mean(),inplace=True)
dv.head(30)
```
